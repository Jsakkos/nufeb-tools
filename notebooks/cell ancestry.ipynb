# Make plots for the documentation


```python
from nufeb_tools import utils,plot
import matplotlib.pyplot as plt
```

```python
x = utils.get_data(directory=r'E:\sucrose\runs\Run_50_50_1.00e+00_1_2022-01-11_48525')
#x = utils.get_data(test=True)
```

```python
f, axes = plt.subplots(ncols=3,figsize=(15,5))

for ax, time in zip(axes,[100,20000,25900]):
    plot.colony(x,time,by='Species',ax=ax)
plt.show()
f.tight_layout()
f.savefig(r'..\docs\_static\images\Colonies_over_time.png')
```

```python
f, ax = plt.subplots(figsize=(8,8))
im = plot.colony(x,25900,by='Species',ax=ax)
#colony(x,25900,ax=ax)
ax.set_axis_off()
f.savefig(r'..\docs\_static\images\colonies.png',dpi=300,bbox_inches='tight')
plt.show()
```

```python
f, ax = plt.subplots(figsize=(8,8))
im = plot.colony(x,20000,by='Species',ax=ax,fitness=True)
ax.set_axis_off()
f.tight_layout()
f.savefig(r'..\docs\_static\images\colony_fitness.png',dpi=300,bbox_inches='tight')
plt.show()
```
